```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf

from matplotlib import rc
rc('font', family='serif')
rc('lines', linewidth=1.5)
rc('font', size=16)
plt.rc('legend',**{'fontsize':11})
```

```python
#DTYPE='float32'
DTYPE='float64'

IB = 4 # Choice of an Initial/Boundary Condition. 0 is our example. 

if IB == 0:
    def fun_v_0(x):
        return -tf.sin(pi * x)   #Also, -np.sin() works    
elif IB == 1:
    def fun_exact(t,x):
        cond1 = tf.cast(tf.math.less_equal(x, t/2.0), dtype=DTYPE)
        cond2 = tf.cast(tf.math.greater(x, t/2.0), dtype=DTYPE)
        a = tf.math.multiply(cond1, 1.0)
        b = tf.math.multiply(cond2, 0.0)#<---Mistake! cond 1 or cond 2?
        return a+b
elif IB == 2:    
    def fun_exact(t,x):
        cond1 = tf.cast(tf.math.less_equal(x, t), tf.float32)
        cond2 = tf.cast(tf.math.logical_and(tf.math.less(x, 1.0), tf.math.greater(x, t)), tf.float32)
        cond3 = tf.cast(tf.math.greater_equal(x, 1), tf.float32)#    
        a = tf.math.multiply(cond1, 1.0)
        b = tf.math.multiply(cond2, (1.0 - x)/(1-t))
        c = tf.math.multiply(cond3, 0.0)
        return a+b+c
elif IB == 3:    
    def fun_exact(t,x):
        cond1 = tf.cast(tf.math.less_equal(x, 0), dtype=DTYPE)
        cond2 = tf.cast(tf.math.logical_and(tf.math.less(x, t), tf.math.greater(x, 0.0)), dtype=DTYPE)
        cond3 = tf.cast(tf.math.greater_equal(x, t), dtype=DTYPE)
        t = tf.where(tf.equal(t, 0), 1e-10 * tf.ones_like(t), t) #WHen t = 0 is a denominator.
        a = tf.math.multiply(cond1, 0.0)
        b = tf.math.multiply(cond2, x/t)
        c = tf.math.multiply(cond3, 1.0)
        return a+b+c
elif IB == 4:    
    def fun_exact(t,x):
        cond00 = tf.cast(tf.math.logical_and(tf.math.less(x, 0.0), tf.math.equal(t, 0.0)), dtype=DTYPE)
        cond01 = tf.cast(tf.math.logical_and(tf.math.less_equal(x, 1), tf.math.greater_equal(x,0.0),
                                             tf.math.equal(t, 0.0)), dtype=DTYPE)
        cond02 = tf.cast(tf.math.logical_and(tf.math.greater(x, 1.0), tf.math.equal(t, 0.0)), dtype=DTYPE)
        
        cond10 = tf.cast(tf.math.logical_and(tf.math.less(x, 0.0), tf.math.greater_equal(t, 0.0),
                                            tf.math.less(t,2.)), dtype=DTYPE)
        cond11 = tf.cast(tf.math.logical_and(
                    tf.math.logical_and(tf.math.less(x, t), tf.math.greater_equal(x, 0.0)), 
                    tf.math.logical_and(tf.math.greater(t, 0.0),  tf.math.less(t,2.0))
                    ), dtype=DTYPE
                 )
        cond12 = tf.cast(tf.math.logical_and(
                    tf.math.logical_and(tf.math.greater_equal(x, t), tf.math.less_equal(x, 1.0+t/2.0)), 
                    tf.math.logical_and(tf.math.greater(t, 0.0),  tf.math.less(t,2.0))
                    ), dtype=DTYPE
                 )
        cond13 = tf.cast(tf.math.logical_and(tf.math.greater(x, 1.0+t/2.0),
                                             tf.math.greater(t, 0.0),  tf.math.less(t,2.0)), dtype=DTYPE)
        
        cond20 = tf.cast(tf.math.logical_and(tf.math.less_equal(x, 0.0), 
                                             tf.math.greater_equal(t, 2.0)), dtype=DTYPE)
        cond21 = tf.cast(tf.math.logical_and(tf.math.greater_equal(x, 0.0), tf.math.less_equal(x, np.sqrt(2.0*t)),
                                             tf.math.greater_equal(t, 2.0)), dtype=DTYPE)
        cond22 = tf.cast(tf.math.logical_and(tf.math.greater(x, np.sqrt(2.0*t)),
                                             tf.math.greater_equal(t, 2.0)), dtype=DTYPE)                                    
       
     #   t = tf.where(tf.equal(t, 0), 1e-10 * tf.ones_like(t), t) #WHen t = 0 is a denominator.--> Not good
      
        a = tf.math.multiply(cond00, 0.0)
        b = tf.math.multiply(cond01, 1.0)
        c = tf.math.multiply(cond02, 0.0)
        
        d = tf.math.multiply(cond10, 0.0)
        e = tf.math.multiply(cond11, x/t)
        f = tf.math.multiply(cond12, 1.0)
        g = tf.math.multiply(cond13, 0.0)
        
        h = tf.math.multiply(cond20, 0.0)
        i = tf.math.multiply(cond21, x/t)
        j = tf.math.multiply(cond22, 0.0)
        
        if t == 0.:
            return a+b+c
        elif  (t>0.0 and t<2.0):
            return d+e+f+g
        else:
            return h+i+j
else:
    print("No more solution?")
```

```python
Nx = 601;                         # Number of grid points
x_max = 8.1;                        # Domain limit to the right
x_min = -2.;                       # Domain limit to the left
dx = (x_max-x_min)/(Nx-1)           # Mesh size
x = np.arange(x_min,x_max,dx)       # Discretized mesh

t_int = 0.0009                      # Initial time
t_end = 12.                    # Final time
dt = 0.01                         # Time step
Nt = int(t_end/dt)                # Number of iterations
t = np.linspace(t_int,t_end,Nt+1)   # Time vector
```

```python
v = np.zeros((t.size, x.size))

#Vectorize the nested for loop in Tensorflow
for i in range(t.size):
    v[i] = fun_exact(t[i],x)
    
print(v.shape)
```

```python
np.save("exact_data4_x_time8", x)
np.save("exact_data4_t_time8", t)
np.save("exact_data4_v_time8", v)
```
